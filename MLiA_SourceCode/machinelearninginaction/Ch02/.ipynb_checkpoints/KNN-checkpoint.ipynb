```python
#-*- coding:utf-8 -*-
from numpy import *
import operator
from os import listdir
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndicies = distances.argsort()                                                    #argsort()
    classCount={}          
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True) #sorted()
    return sortedClassCount[0][0]
```

numpy.argsort():对列表进行正序排序后，返回排序后每个元素在原列表中的位置。  
sorted():返回列表。  
tile(inX, (n, m)):复制为n.m的矩阵  
**构建K近邻分类器,只处理一条测试数据**  
    -input:[testData list, trainData array, trainLabels list, NumOfNearst int]  
    -output:[labelOf_testData char]  

```python
def createDataSet():
    group = array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]])
    labels = ['A','A','B','B']
    return group, labels
```

**创建数据和标签**  
-input:  
-output:[trianData array, labels list]  

```python
inX = [0, 0]
trainData, trainLabel = createDataSet()
numOfNearst = 3
print classify0(inX, trainData, trainLabel, numOfNearst)
```

```python
def classify1(testData, trainData, trainLabel, k):
    returnLabelList = []
    for inX in testData:
        returnLabelList.append(classify0(inX, trainData, trainLabel, k))
    return returnLabelList
```

**在classify0()上，对一组测试数据进行分类**  
    -input：[testData array, trainData array, trainLabel list, k int]  
    -output：[returnLabelList list]

```python
def file2matrix(filename):
    fr = open(filename)
    numberOfLines = len(fr.readlines())
    returnMat = zeros((numberOfLines, 3))
    classLabelVector = []
    fr = open(filename)
    index = 0
    for line in fr.readlines():
        line = line.strip()                          #strip()
        listFromLine = line.split()                  #split()
        returnMat[index, :] = listFromLine[0: 3]
        if listFromLine[-1] == 'largeDoses':
            label = 3
        elif listFromLine[-1] == 'smallDoses':
            label = 2
        elif listFromLine[-1] == 'didntLike':
            label = 1
        else :
            label = 0
        classLabelVector.append(label)
        index += 1
    return returnMat, classLabelVector
```

注意：文件读到尾后，需要重新打开，否则读出内容为空。

```python
def autoNorm(dataSet):
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    normDataSet = zeros(shape(dataSet))
    m = dataSet.shape[0]
    normDataSet = dataSet - tile(minVals, (m, 1))
    normDataSet = normDataSet / tile(ranges, (m, 1))
    return normDataSet, ranges, minVals
```

    -input：[dataSet array]  
    -output:[normDataSet array, ranges array,minVals array]

```python
def datingClassTest():
    hoRatio = .1
    datingDataMat, datingLabels = file2matrix('datingTestSet.txt')
    normMat, ranges, minVals = autoNorm(datingDataMat)
    m = normMat.shape[0]
    numTestVecs = int(m*hoRatio)
    errorCount = .0
    for i in range(numTestVecs):
        classifierResult = classify0(normMat[i, :], normMat[numTestVecs:m, :], datingLabels[numTestVecs:m], 3)
        #print 'the classifier came back with: %d, the real answer is: %d' % (classifierResult, datingLabels[i]) 
        if(classifierResult != datingLabels[i]):
            errorCount += 1.
    print 'the total error is: %f' % (errorCount/float(numTestVecs))
    print errorCount
```

```python
datingClassTest()
# 约会测试，选取前100进行测试，给出正确率
```

```python
def datingPerson():
    resultList = ['not at all', 'in samll doses', 'in large doses']
    a = float(raw_input("frequent flier miles earned per year:"))
    b = float(raw_input("percentage of time spent playing video game:"))
    c = float(raw_input("liters of ice cream consumed per week:"))
    datingMat, datingLabels = file2matrix('datingTestSet.txt')
    normMat, ranges, minVals = autoNorm(datingMat)
    inX = array([a, b, c])
    norm_inX = (inX - minVals) / ranges
    label = classify0(norm_inX, normMat, datingLabels, 3)
    print 'you will probably like this person ' + resultList[label - 1] + ' !'
# 交互，可输入自己的数据进行归类
```

```python
def img2vector(filename):
    returnVect = zeros((1, 1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32*i+j] = int(lineStr[j])
    return returnVect
```

```python
#filename = '../../data/digits/testDigits/0_0.txt'
filename = 'digits/testDigits/0_0.txt'
img2vector(filename)
```

```python
def handwritingClassTest():
    hwLabels = []
    trainingFileList = listdir('digits/trainingDigits')
    m = len(trainingFileList)
    trainingMat = zeros((m, 1024))
    for i in range(m):
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        hwLabels.append(classNumStr)
        trainingMat[i, :] = img2vector('digits/trainingDigits/%s' % fileNameStr)
    
    testFileList = listdir('digits/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        vectorUnderTest = img2vector('digits/testDigits/%s' % fileNameStr)
        classifierResult = classify0(vectorUnderTest, trainingMat, hwLabels, 3)
        #print 'the classifier came back with: %d, the real answer is: %d' % (classifierResult, classNumStr)
        if(classifierResult != classNumStr):
            errorCount += 1.
    print '\nthe total number of errors is: %d' % errorCount
    print '\nthe total error rate is: %f' % (errorCount / float(mTest))
```

```python
handwritingClassTest()
```

**K近邻算法总结（有导师学习）**  
k近邻算法是分类数据最简单有效的算法；  
k近邻算法步骤：  
1.准备数据，文本信息到方便处理的矩阵array，和标签信息提取；  
2.分析数据，哪些数据适合分类使用；  
3.规格数据；  
4.分类，k值的选取；  
  
**缺点：**  
1.必须保存所有的数据集，如果训练数据集很大，需要占据大量的存储空间；  
2.O(NM)距离计算十分耗时；  
3.无法给出任何数据的基础结构信息，我们无法知道平均实例样本和典型样本具有什么样的特征；  
  
**k决策树**  
是k近邻算法的优化版；  
能减少存储空间和计算时间开销；
